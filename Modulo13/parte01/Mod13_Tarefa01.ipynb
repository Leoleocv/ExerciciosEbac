1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.  


2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.  


3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python

y, x = patsy.dmatrices( 'np.log(renda) ~ C(tipo_renda) + C(educacao) + C(estado_civil) + C(tipo_residencia)+ qtd_filhos + posse_de_veiculo + posse_de_imovel+ idade + tempo_emprego + qt_pessoas_residencia', data=df, return_type='dataframe')

x
```

```python
sm.OLS(y,x).fit().summary()
```

Agora colocando as variaveis mais frequentes como casela de referencia.

```python
x.sum(axis=0)
```

As categorias mais frequentes das variaveis qualitativas sao: 'C(tipo_renda)[T.Empresário]' , 'C(educacao)[T.Secundário]', 'C(estado_civil)[T.Solteiro]', 'C(tipo_residencia)[T.Casa]'.

```python

y, x = patsy.dmatrices('np.log(renda) ~ C(tipo_renda, Treatment("Empresário")) + C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Solteiro")) + C(tipo_residencia, Treatment("Casa")) + qtd_filhos + posse_de_veiculo + posse_de_imovel+ idade + tempo_emprego + qt_pessoas_residencia', data=df, return_type='dataframe')

sm.OLS(y, x).fit().summary()
```

Removendo a variavel C(tipo_renda, Treatment("Empresário"))[T.Bolsista] , a qual possui p = 0.918.

```python
x = x.drop('C(tipo_renda, Treatment("Empresário"))[T.Bolsista]', axis=1)

sm.OLS(y, x).fit().summary()
```

Apos ser removido a categoria T.Bolsista o R ajustado do modelo aumentou de 0.206 para 0.207 mostrando que mesmo que muito pequena, a remoção desta variavel foi benéfica, fora essas mudanças nao houve uma mudança significativa nos parâmetros.

A com maior p agora é a variavel C(estado_civil, Treatment("Solteiro"))[T.Separado]

```python
x = x.drop('C(estado_civil, Treatment("Solteiro"))[T.Separado]', axis=1)

sm.OLS(y, x).fit().summary()
```

A com maior p agora é a variavel C(educacao, Treatment("Secundário"))[T.Pós graduação]

```python
x = x.drop('C(educacao, Treatment("Secundário"))[T.Pós graduação]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(educacao, Treatment("Secundário"))[T.Superior incompleto]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(tipo_residencia, Treatment("Casa"))[T.Governamental]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(educacao, Treatment("Secundário"))[T.Primário]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(tipo_residencia, Treatment("Casa"))[T.Comunitário]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(tipo_renda, Treatment("Empresário"))[T.Pensionista]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('C(estado_civil, Treatment("Solteiro"))[T.Casado]', axis=1)

sm.OLS(y, x).fit().summary()
```

```python
x = x.drop('qtd_filhos', axis=1)

sm.OLS(y, x).fit().summary()
```

-Modelo inicial: R-squared = 0.208, Adj. R-squared = 0.207.

-Modelo final: R-squared = 0.207, Adj. R-squared = 0.206.

A capacidade explicativa do modelo final é praticamente igual à do modelo inicial, indicando que as variáveis removidas tinham pouca ou nenhuma contribuição explicativa.


-Inicial: AIC = 29780.

-Final: AIC = 29770.

Ambos o critério melhororou levemente no modelo final, indicando maior eficiência do modelo em relação a troca entre ajuste e número de variáveis.
