```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize = ( 8, 6))

sns.boxplot(x = df['IDADEMAE'])
plt.title('Boxplot da idade da mãe')
plt.xlabel('idade da mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = df['PESO'])
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins = 20, kde = True , color = 'skyblue')
plt.title('Histograma da idade da mãe')
plt.xlabel('Idade da mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['PESO'], bins = 20, kde = True, color = 'salmon')
plt.title('Histograma do peso do Bebe')
plt.xlabel('Peso do Bebe (gramas)')
plt.ylabel('Frequência')
plt.show()
```
