### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf

import numpy as np
````

````python
tips = sns.load_dataset('tips')
````

````python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
reg = smf.ols('tip_pct ~ net_bill', data = tips).fit()
reg.summary()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
reg = smf.ols('tip_pct ~ np.log(net_bill)', data = tips).fit()
reg.summary()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
reg = smf.ols('tip_pct ~ net_bill + np.power(net_bill, 2)', data = tips).fit()
reg.summary()
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
reg = smf.ols('np.log(tip_pct) ~ net_bill', data = tips).fit()
reg.summary()
````

````python
tip_pct_pred_log = reg.fittedvalues
tip_pct_pred = np.exp(tip_pct_pred_log)
SST = np.sum((tips['tip_pct'] - tips['tip_pct'].mean())**2)  # Soma total dos quadrados
SSR = np.sum((tips['tip_pct'] - tip_pct_pred)**2)            # Soma dos quadrados dos resíduos

R2_tip_pct = 1 - (SSR / SST)
R2_tip_pct
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
reg2 = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data = tips).fit()
reg2.summary()
````

````python
tip_pct_pred_log2 = reg2.fittedvalues
tip_pct_pred2 = np.exp(tip_pct_pred_log2)

SST2 = np.sum((tips['tip_pct'] - tips['tip_pct'].mean())**2)  # Soma total dos quadrados
SSR2 = np.sum((tips['tip_pct'] - tip_pct_pred2)**2)            # Soma dos quadrados dos resíduos

R2_tip_pct2 = 1 - (SSR2 / SST2)
R2_tip_pct2
````

###  6. qual modelo resultou no melhor ajuste?

O ultimo modelo onde foi ajustado no log de do tip_pct em função do log de net_bill, pois obteve o maior coeficiente de determinação com 0.3219
