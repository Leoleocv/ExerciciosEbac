#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
apple = apple.drop(index=apple.index[:2])
apple = apple.rename(columns={'Price': 'Date'})

amazon = amazon.drop(index=amazon.index[:2])
amazon = amazon.rename(columns={'Price': 'Date'})

microsoft = microsoft.drop(index=microsoft.index[:2])
microsoft = microsoft.rename(columns={'Price': 'Date'})
````

````python
amazon.head()
````

````python
apple['Mmovel90'] = apple['Close'].rolling(window=90).mean()
amazon['Mmovel90'] = amazon['Close'].rolling(window=90).mean()
microsoft['Mmovel90'] = microsoft['Close'].rolling(window=90).mean()
````

````python
apple['STD90'] = apple['Close'].rolling(window=90).std()
amazon['STD90'] = amazon['Close'].rolling(window=90).std()
microsoft['STD90'] = microsoft['Close'].rolling(window=90).std()
````

````python
fig = px.line(apple, x='Date', y='Mmovel90', title='Média Móvel 90 dias - Apple', labels={'Mmovel90': 'Média Móvel', 'Date' : 'Data'})
fig.show()
````

````python
fig = px.line(amazon, x='Date', y='Mmovel90', title='Média Móvel 90 dias - Amazon', labels={'Mmovel90': 'Média Móvel', 'Date' : 'Data'})
fig.show()
````

````python
fig = px.line(microsoft, x='Date', y='Mmovel90', title='Média Móvel 90 dias - Microsoft', labels={'Mmovel90': 'Média Móvel', 'Date' : 'Data'})
fig.show()
````

````python
fig = px.line(apple, x='Date', y='STD90', title='Desvio Padrão 90 dias - Apple', labels={'STD90': 'Desvio Padrão', 'Date' : 'Data'})
fig.show()

fig2 = px.line(amazon, x='Date', y='STD90', title='Desvio Padrão 90 dias - Amazon', labels={'STD90': 'Desvio Padrão', 'Date' : 'Data'})
fig2.show()

fig3 = px.line(microsoft, x='Date', y='STD90', title='Desvio Padrão 90 dias - Microsoft', labels={'STD90': 'Desvio Padrão', 'Date' : 'Data'})
fig3.show()
````

Olhando para os gráficos é possivel concluir que o ativo com maior volatilidade foi o da microsoft apresentando maior desvio padrao alem de oscilações nos periodos de alta e baixa volatilidade.Em todos os casos, os períodos de maior volatilidade geralmente coincidiram com mudanças significativas na tendência de preço ou com acelerações em tendências existentes. A Microsoft apresentou a maior volatilidade absoluta entre os três ativos, enquanto também manteve o maior valor médio de preço. A Apple apresentou o movimento mais dramático, com volatilidade alta acompanhando um aumento de aproximadamente 30% no preço médio entre junho e dezembro de 2024. A volatilidade de todos os ativos mostrou comportamento cíclico, indicando períodos alternados de estabilidade e instabilidade no mercado.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
import plotly.graph_objects as go

microsoft_last60 = microsoft.tail(60)

fig = go.Figure(data=[go.Candlestick(
    x=microsoft_last60.index,
    open=microsoft_last60['Open'],
    high=microsoft_last60['High'],
    low=microsoft_last60['Low'],
    close=microsoft_last60['Close'],
    name='Microsoft (Últimos 60 dias)'
)])

fig.update_layout(
    title='Gráfico Candlestick - Microsoft (Últimos 60 dias)',
    paper_bgcolor='white',
    plot_bgcolor='white',
    xaxis=dict(
        showgrid=True,
        gridcolor='lightgrey',
        rangeslider=dict(visible=False)
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='lightgrey',
        title='Preço (USD)'
    )
)

fig.show()
````
