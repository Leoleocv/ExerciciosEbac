#### 1. Baixando dados de ativos
1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import os
```

```python
if not os.path.exists('output'):
    os.makedirs('output')
```

```python
end_date = datetime.now()
start_date = end_date - timedelta(days=2*365)

ativos = ['AAPL', 'MSFT', 'AMZN']  # Apple, Microsoft e Amazon

data = {}
for ativo in ativos:
    data[ativo] = yf.download(ativo, start=start_date, end=end_date)

    csv_path = f'./output/{ativo}_dados.csv'
    data[ativo].to_csv(csv_path)
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
for ativo in ativos:
    data[ativo]['amplitude'] = data[ativo]['High'] - data[ativo]['Low']
```

```python
nomes = ['Apple', 'Microsoft', 'Amazon']

for i, ativo in enumerate(ativos):
    plt.figure(figsize=(10, 6))
    plt.plot(data[ativo].index, data[ativo]['amplitude'], label= nomes[i], alpha=0.7)

    plt.title(f'Amplitude de Variação Diária ({ativo})', fontsize=16)
    plt.xlabel('Data', fontsize=12)
    plt.ylabel('Amplitude em USD', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()  
    plt.show() 
```

Olhando para os gráficos de variação de amplitudes, as ações da amazon e da microsoft possuem mudanças mais bruscas que ações da apple, com a amazon tendo as mudanças mais bruscas indicando ser a mais volatil das 3.
